# Fetching Datasets

```python
df1=pd.read_csv("/content/prevalence-by-mental-and-substance-use-disorder (1).csv")

df2=pd.read_csv("/content/mental-and-substance-use-as-share-of-disease.csv")
```

```python
data = pd.merge(df1, df2) #Merging datasets
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.drop('Code', axis = 1, inplace = True) #Droping non-usable column.
```

```python
data.size, data.shape
```

```python
data.head(2)
```

```python
#Renaming columns name
data.set_axis(['Country', 'Year', 'Schizophrenia', 'Bipolar_disorder', 'Eating_disorder', 'Anxiety', 'Drug_use', 'Depressive_disorder','alcohol','mental_fitness'], axis = 'columns', inplace = True)
```

```python
df = data
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()

df['Country'] = l.fit_transform(df['Country'])
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('mental_fitness', axis = 1)
y = df['mental_fitness']

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2, random_state = 2)
```

```python
print("xtrain : ", xtrain.shape)
print("xtest : ", xtest.shape)
print("ytrain : ", ytrain.shape)
print("ytest : ", ytest.shape)
```

# Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(xtrain, ytrain)

ytrain_pred = lr.predict(xtrain)

mse = mean_squared_error(ytrain, ytrain_pred)

rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

r2 = r2_score(ytrain, ytrain_pred)


print("The Linear Regression model performance for training set")
print("--------------------------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score {}'.format(r2))
```

# Applying Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)

mse = mean_squared_error(ytrain, ytrain_pred)

rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

r2 = r2_score(ytrain, ytrain_pred)


print("The RandomForest Regression model performance for training set")
print("--------------------------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score {}'.format(r2))
```

# Applying Decission Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(random_state=0)
dtr.fit(xtrain,ytrain)

ytest_pred=dtr.predict(xtest)

from sklearn.metrics import r2_score,mean_squared_error
print("Results for Decision Tree Regression:\nMean Square Error={}\nR-Square Score={}"
      .format(mean_squared_error(ytest,ytest_pred),r2_score(ytest,ytest_pred)))
```

# SVM Regression

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(xtrain,ytrain)

ytest_pred=svr.predict(xtest)

from sklearn.metrics import r2_score,mean_squared_error
print("Results for SVM Regression:\n Mean Square Error={}\n R-Square Score={}".format(mean_squared_error(ytest,ytest_pred),r2_score(ytest,ytest_pred)))
```

# Final Result

Random Forest Regression demonstrates excellent performance on both the training and testing sets, achieving an impressive R-squared score of 0.99. Similarly, Decision Tree Regression also exhibits strong performance on both the training and testing sets, yielding a notable R-squared score of 0.98.
